```python
from PIL import Image
from PIL import ImageOps 
from matplotlib import pyplot as plt
import numpy as np
import cv2


image= Image.open('dog.jpg')
plt.imshow(image)
plt.show()
```

```python
print(image.mode)
print(image.size)
print(image.format)
```

```python

image_gray = ImageOps.grayscale(image)
image_gray
```

```python
image_gray.mode
```

```python
image_gray.quantize(256//9)
image_gray
```

```python
image_array = np.array(image)
plt.figure(figsize=(10,10))
plt.imshow(image_array)
plt.show()
```

```python
image_red=image_array.copy()
image_red[:,:,1] = 0
image_red[:,:,2] = 0
plt.figure(figsize=(10,10))
plt.imshow(image_red)
plt.show()
```

```python
image_blue=image_array.copy()
image_blue[:,:,0] = 0
image_blue[:,:,2] = 0
plt.figure(figsize=(10,10))
plt.imshow(image_blue)
plt.show()
```

```python
im_flip=ImageOps.flip(image)
plt.imshow(im_flip)
plt.show()
```

```python
im_mirror=ImageOps.mirror(image)
plt.imshow(im_mirror)
plt.show()
```

```python
tom_flip = cv2.rotate(np.array(tom), cv2.ROTATE_90_CLOCKWISE)
plt.imshow(tom_flip)
plt.show()
```
